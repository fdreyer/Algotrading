```python
import requests
import pprint
assets = requests.get(url='https://api.tinychart.org/assets/')
prices = requests.get(url='https://api.tinychart.org/prices/')
tiny = requests.get(url='https://api.tinychart.org/asset/378382099/')
tinyprice = requests.get(url='https://api.tinychart.org/asset/378382099/price/')
tinyprices = requests.get(url='https://api.tinychart.org/asset/378382099/prices/?start=1636270738')
```

```python
print('Assets:')
pprint.pprint(assets.json())
```

```python
print('Prices:')
pprint.pprint(prices.json())
```

```python
print('Tiny token:')
pprint.pprint(tiny.json())
print('Tiny price:')
pprint.pprint(tinyprice.json())
```

```python
print('Tiny historical prices')
pprint.pprint(tinyprices.json())
```
